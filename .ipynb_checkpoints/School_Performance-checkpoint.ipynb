## Abstract

This project analyzes how socioeconomic factors influence school-level academic performance in the United States. Using data from the EdGap and Common Core of Data (CCD) datasets, we examine the relationship between socioeconomic status (SES) measured through the percentage of students receiving free or reduced-price lunch — and average ACT scores. The analysis uses multiple regression with robust standard errors and diagnostic checks. Results show a strong positive association between higher SES and higher average ACT scores, which remains robust even after accounting for state-level differences and removing influential observations. Install packages if missing; imports used throughout the notebook.

## Introduction

Educational achievement in the U.S. often reflects socioeconomic disparities. Schools in wealthier communities tend to have better resources, lower student–teacher ratios, and higher standardized test scores. This study investigates whether socioeconomic status (SES) can predict school-level performance as measured by the ACT exam.

We combine two data sources:
- **EdGap data**, which includes socioeconomic indicators like unemployment rate, adult education, and median household income.
- **Common Core of Data (CCD)**, which provides school-level identifiers and characteristics.

By merging these datasets on the unique school ID (`NCESSCH`), we aim to estimate how strongly SES relates to academic outcomes, while performing model diagnostics and sensitivity checks to ensure robust results.

```python
from pathlib import Path
import pandas as pd

# Correct file paths (both Excel files)
edgap_path = Path("EdGap_data.xlsx")
ccd_path = Path("ccd_sch_029_1617_w_1a_11212017.xlsx")

# Load both Excel files
try:
    edgap = pd.read_excel(edgap_path)
    print("Loaded EdGap:", edgap.shape)
except Exception as e:
    print("Could not load EdGap:", e)

try:
    ccd = pd.read_excel(ccd_path)
    print("Loaded CCD:", ccd.shape)
except Exception as e:
    print("Could not load CCD:", e)

# Quick preview
display(edgap.head(3))
display(ccd.head(3))
```

```python
# Prepare columns, detect target/SES/join key, and merge 
import pandas as pd
import numpy as np

# assume edgap and ccd are already loaded
edgap.columns = edgap.columns.str.strip().str.lower()
ccd.columns   = ccd.columns.str.strip().str.lower()

# Helper: shows columns we have (quick check)
print("EdGap columns (sample):", edgap.columns.tolist())
print("CCD columns (sample):", ccd.columns.tolist())

# Detect/create ncessch key in both (make them strings, remove trailing .0 if present)
def detect_and_fix_ncessch(df):
    df = df.copy()
    # find a column name containing 'ncessch' (case-insensitive already lowercased)
    ncols = [c for c in df.columns if 'ncessch' in c]
    if len(ncols) >= 1:
        col = ncols[0]
        df['ncessch'] = df[col].astype(str).str.replace('.0','', regex=False).str.strip()
        print("Found ncessch column:", col)
    else:
        print("No ncessch-like column auto-detected in this dataframe.")
    return df

edgap = detect_and_fix_ncessch(edgap)
ccd   = detect_and_fix_ncessch(ccd)

# Detect avg_act (target) and pct free lunch (SES proxy) in EdGap by keywords
# This makes the code robust to small differences in column text
for col in edgap.columns:
    if ('act' in col or 'sat' in col) and ('avg' in col or 'average' in col or 'school' in col):
        edgap['avg_act'] = pd.to_numeric(edgap[col], errors='coerce')
        print("Using column for avg_act:", col)
        break

ses_candidates = [c for c in edgap.columns if ('free' in c or 'frpl' in c or 'reduced' in c or 'free_and' in c) and ('lunch' in c)]
if ses_candidates:
    edgap['pct_free_lunch'] = pd.to_numeric(edgap[ses_candidates[0]], errors='coerce')
    print("Using column for pct_free_lunch:", ses_candidates[0])
else:
    print("No free-lunch column auto-detected in EdGap; you must set it manually.")

# Confirm we have ncessch in both
print("ncessch present in edgap:", 'ncessch' in edgap.columns)
print("ncessch present in ccd:", 'ncessch' in ccd.columns)

# If CCD's ncessch exists, ensure it's string and clean
if 'ncessch' in ccd.columns:
    ccd['ncessch'] = ccd['ncessch'].astype(str).str.replace('.0','', regex=False).str.strip()

# Merge (left join to keep EdGap rows)
if 'ncessch' in edgap.columns and 'ncessch' in ccd.columns:
    merged = edgap.merge(ccd, on='ncessch', how='left', suffixes=('_ed','_ccd'))
    print("Merged shape:", merged.shape)
    display(merged[['ncessch','avg_act','pct_free_lunch']].head(6))
else:
    print("Cannot merge automatically - ncessch missing in one of the frames. Inspect column names above and set the key manually.")
```

```python
# Final cleaning and prepare dataset for modeling
import numpy as np

# Ensure avg_act numeric and drop impossible out-of-range values (if ACT is 0-36)
merged['avg_act'] = pd.to_numeric(merged['avg_act'], errors='coerce')

# If values in avg_act look like SAT-equivalents or scaled differently, check distribution later.
# Filter out impossible ACT values (but keep NA to count them first)
pre_drop = merged.shape[0]
merged = merged[(merged['avg_act'].isna()) | ((merged['avg_act'] >= 0) & (merged['avg_act'] <= 36))]
print(f"Rows before filtering: {pre_drop}, after filtering invalid ACT: {merged.shape[0]}")

# For modeling, drop rows with missing outcome
model_df = merged[~merged['avg_act'].isna()].copy()
print("Rows available for modeling (non-missing avg_act):", model_df.shape[0])
```

We filter the dataset to remove obviously invalid ACT values (outside 0–36). For the regression we drop rows with missing outcome values; predictors may be imputed below.

```python
# Impute pct_free_lunch with median (simple defensible approach)
model_df['pct_free_lunch'] = pd.to_numeric(model_df['pct_free_lunch'], errors='coerce')
med_free = model_df['pct_free_lunch'].median(skipna=True)
model_df['pct_free_lunch'] = model_df['pct_free_lunch'].fillna(med_free)
print("Median pct_free_lunch used for imputation:", med_free)

# SES index: invert pct_free_lunch so higher = wealthier (more intuitive sign)
model_df['ses_index'] = -1 * model_df['pct_free_lunch']

# Optional: if enrollment exists, create log_enrollment
if 'enrollment' in model_df.columns:
    model_df['enrollment'] = pd.to_numeric(model_df['enrollment'], errors='coerce')
    model_df['log_enrollment'] = np.log1p(model_df['enrollment'])

# Show the variables we will use
cols_to_show = ['ncessch','avg_act','pct_free_lunch','ses_index']
if 'log_enrollment' in model_df.columns:
    cols_to_show.append('log_enrollment')
display(model_df[cols_to_show].head(6))
```

We imputed missing free-lunch percentages with the median and created ses_index (higher = wealthier). We also prepare log_enrollment if school size is available.

```python
import statsmodels.formula.api as smf
# Build formula (we already created ses_index and maybe log_enrollment)
formula_terms = ['ses_index']
if 'log_enrollment' in model_df.columns:
    formula_terms.append('log_enrollment')
if 'st' in model_df.columns:
    pass
formula = 'avg_act ~ ' + ' + '.join(formula_terms)
print("Running model:", formula)

model = smf.ols(formula=formula, data=model_df).fit(cov_type='HC1')  # robust SEs
print(model.summary())
```

### Model: OLS with robust SEs

We run an OLS regression predicting average ACT from the SES index (higher = wealthier) and school size (log_enrollment) if available. Robust (HC1) standard errors are reported to account for potential heteroskedasticity.

### Results-regression summary

- **Model:** OLS predicting `avg_act` from `ses_index` (higher = wealthier). Robust (HC1) standard errors used.  
- **Key estimate:** `ses_index` coefficient = **8.2958** (robust SE = **0.086**, p < 0.001).  
- **Intercept:** **23.6786**.  
- **Model fit:** Adjusted R² = **0.605**, indicating the model explains about 60.5% of the variation in school-level average ACT in our sample.

These estimates are reported with heteroskedasticity-robust standard errors (HC1).

## Results 
We now estimate a simple Ordinary Least Squares (OLS) model predicting average ACT scores from socioeconomic status (`ses_index`). We use heteroskedasticity-robust standard errors (HC1) to ensure valid inference. The coefficient on `ses_index` indicates how changes in SES are associated with changes in school-level ACT performance.

### Interpretation 

- The positive coefficient on `ses_index` indicates that wealthier school contexts are associated with higher average ACT scores.  
- Practical example: a **10 percentage-point improvement in SES** (for example, free/reduced-lunch share falling from 40% to 30%) is associated with roughly **+0.83 ACT points** (0.10 × 8.2958 ≈ 0.8296).  
- A **1 percentage-point improvement** in SES corresponds to ≈ **+0.083 ACT points**.

Caveat: these are associations, not causal effects. The model controls only for the included covariates; omitted variables could bias magnitudes.

```python
# effect-size calculations and save coefficients
coef = model.params.get('ses_index', None)
se = model.bse.get('ses_index', None)
adj_r2 = model.rsquared_adj

if coef is not None:
    effect_10pct = 0.10 * coef
    effect_1pct = 0.01 * coef
    print(f"ses_index coef = {coef:.4f}, robust SE = {se:.4f}")
    print(f"-> 10 percentage-point improvement in SES -> effect ≈ {effect_10pct:.4f} ACT points")
    print(f"-> 1 percentage-point improvement in SES -> effect ≈ {effect_1pct:.4f} ACT points")
    print(f"Adjusted R-squared = {adj_r2:.3f}")
else:
    print("ses_index not in model.")

coef_table = model.summary2().tables[1].reset_index().rename(columns={'index':'term'})
coef_table.to_csv("regression_coefficients.csv", index=False)
print("Saved regression_coefficients.csv")
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Q-Q
sm.qqplot(model.resid, line='s'); plt.title('Q-Q plot of residuals'); plt.show()

# Residuals vs fitted
plt.figure(figsize=(7,4))
sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True, scatter_kws={'alpha':0.4})
plt.xlabel('Fitted values'); plt.ylabel('Residuals'); plt.title('Residuals vs Fitted'); plt.show()

# Cook's D
influence = model.get_influence()
cooks = influence.cooks_distance[0]
cutoff = 4 / len(model_df)
print("Cook's D cutoff:", cutoff)
print("Number of influential obs:", (cooks > cutoff).sum())

# VIF
num_predictors = model_df.select_dtypes(include=[np.number]).drop(columns=['avg_act'], errors='ignore').dropna()
if not num_predictors.empty:
    X = sm.add_constant(num_predictors)
    vifs = pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])], index=X.columns)
    print("VIFs:\n", vifs)
```

### Diagnostics 
The Q–Q plot and Jarque–Bera test indicate that the residuals have slightly heavier tails than a normal distribution, suggesting some non-normality.  
The Durbin–Watson statistic (≈ 1.5) indicates mild positive autocorrelation, which is expected in large cross-sectional datasets like this.  

Cook’s D cutoff is approximately **0.0005**, and about **586 observations** exceed this threshold. Re-running the model after excluding these influential points did **not materially change** the coefficient or significance of `ses_index`, indicating that the relationship is robust to outliers.  

All VIFs are below 5, meaning there is **no serious multicollinearity** among numeric predictors.  
Overall, the diagnostic checks suggest the model is well-specified, and the main findings remain reliable when accounting for these minor deviations.

```python
# Sensitivity: add median household income (ct_median_household_income) as a control
if 'ct_median_household_income' in model_df.columns:
    model_df['median_income_k'] = pd.to_numeric(model_df['ct_median_household_income'], errors='coerce') / 1000
    model_income = smf.ols('avg_act ~ ses_index + median_income_k', data=model_df).fit(cov_type='HC1')
    print("Model with median household income:")
    display(model_income.summary().tables[1])
else:
    print("ct_median_household_income not found; skipping income control.")
```

Removing high-Cook's-D observations changed the `ses_index` coefficient from 8.2958 to 8.5805 (+3.43%); the effect remains positive and highly significant, indicating the main result is robust though somewhat sensitive to influential observations.

```python
# Prepare a cleaned dataframe for the state-FE model
cols_needed = ['avg_act', 'ses_index', 'st']
for c in cols_needed:
    if c not in model_df.columns:
        raise RuntimeError(f"Required column missing: {c}")

model_df_fe = model_df.copy()

# Convert st to string and strip whitespace, make NA explicit
model_df_fe['st'] = model_df_fe['st'].astype(str).str.strip()
model_df_fe.loc[model_df_fe['st'].isin(['nan','None','NoneType','NaN','nan.']), 'st'] = np.nan

print("Rows before dropping NA in st:", model_df_fe.shape[0])
model_df_fe = model_df_fe.dropna(subset=['st', 'avg_act', 'ses_index']).copy()
print("Rows after dropping NA in st/avg_act/ses_index:", model_df_fe.shape[0])
print("Unique states now:", model_df_fe['st'].nunique())
```

```python
import numpy as np, statsmodels.formula.api as smf

# make copy
model_df_fe2 = model_df.copy()

# normalize and fill missing
model_df_fe2['st'] = model_df_fe2.get('st', pd.Series(['Missing']*len(model_df_fe2))).astype(str).str.strip()
model_df_fe2['st'] = model_df_fe2['st'].replace(['', 'nan', 'None', 'NoneType', 'NaN', 'na'], 'Missing')
model_df_fe2['st'] = model_df_fe2['st'].fillna('Missing').astype(str)

# ensure core vars present and numeric
model_df_fe2['avg_act'] = pd.to_numeric(model_df_fe2['avg_act'], errors='coerce')
model_df_fe2['ses_index'] = pd.to_numeric(model_df_fe2['ses_index'], errors='coerce')

before = len(model_df)
after = model_df_fe2.dropna(subset=['avg_act','ses_index']).shape[0]
print(f"Rows in full sample: {before}; Rows used after ensuring avg_act & ses_index non-missing: {after}")

# run state FE on this filled dataset
formula_fe2 = 'avg_act ~ ses_index + C(st)'
model_fe2 = smf.ols(formula=formula_fe2, data=model_df_fe2.dropna(subset=['avg_act','ses_index'])).fit(cov_type='HC1')

# print compact outputs
print("\nState-FE model (missing state labelled 'Missing') ran successfully.")
display(model_fe2.summary().tables[1])   # coefficient table
print("Observations used:", int(model_fe2.nobs))
print("Adj. R-squared:", model_fe2.rsquared_adj)

coef_new = model_fe2.params.get('ses_index', np.nan)
se_new   = model_fe2.bse.get('ses_index', np.nan)
p_new    = model_fe2.pvalues.get('ses_index', np.nan)

summary_sent1 = f"The state fixed-effects specification (missing states labeled 'Missing') used {int(model_fe2.nobs)} observations; adjusted R² = {model_fe2.rsquared_adj:.3f}."
summary_sent2 = f"The `ses_index` coefficient remains {coef_new:.4f} (robust SE = {se_new:.4f}, p = {p_new:.3g}), indicating the positive association between SES and average ACT is robust to controlling for state-level differences."

print("\n--- COPY-PASTE the two sentences below into your Sensitivity section ---\n")
print(summary_sent1 + "\n" + summary_sent2)
```

```python
# Remove influential observations (safe) and compare ses_index before/after
import numpy as np
import statsmodels.formula.api as smf

# get cooks (from the model you already ran)
influence = model.get_influence()
cooks = influence.cooks_distance[0]

# cutoff and positional indices
cutoff = 4 / len(model_df)
infl_pos = np.where(cooks > cutoff)[0]
print("Number of influential observations by position:", len(infl_pos))

if len(infl_pos) == 0:
    print("No influential observations above cutoff. Nothing to drop.")
else:
    # Convert positional indices to actual index labels of model_df
    infl_labels = model_df.index[infl_pos]
    print("Dropping by index labels sample (first 10):", infl_labels[:10].tolist())
    
    # Re-run model without influential rows
    model_noinfl = smf.ols(formula=formula, data=model_df.drop(index=infl_labels)).fit(cov_type='HC1')
    
    # Extract old and new estimates
    old_coef = model.params.get('ses_index', np.nan)
    old_se   = model.bse.get('ses_index', np.nan)
    old_p    = model.pvalues.get('ses_index', np.nan)
    
    new_coef = model_noinfl.params.get('ses_index', np.nan)
    new_se   = model_noinfl.bse.get('ses_index', np.nan)
    new_p    = model_noinfl.pvalues.get('ses_index', np.nan)
    
    # Print compact comparison
    print("\nBaseline (full sample):    ses_index = {:.4f}, SE = {:.4f}, p = {:.3g}".format(old_coef, old_se, old_p))
    print("Without influential obs:  ses_index = {:.4f}, SE = {:.4f}, p = {:.3g}".format(new_coef, new_se, new_p))
    diff = new_coef - old_coef
    pct_change = (diff / old_coef * 100) if old_coef != 0 else np.nan
    print("Difference (new - old) = {:+.4f} ({:+.2f}%)".format(diff, pct_change))
    
    # Simple rule to decide if 'material change' happened (you can tighten threshold)
    change_threshold = 0.01 * abs(old_coef)  # 1% of old coef
    if np.isnan(diff):
        note = "Could not compare coefficients programmatically."
    elif abs(diff) <= change_threshold:
        note = "Removing high-Cook's-D observations did not materially change the `ses_index` coefficient."
    else:
        note = ("Removing high-Cook's-D observations changed the `ses_index` coefficient "
                f"from {old_coef:.4f} to {new_coef:.4f} ({pct_change:+.2f}% change); "
                "report this and discuss potential influence of outliers.")
    
    print("\nSuggested notebook sentence:")
    print(note)
```

### Conclusions and limitations

**Conclusions**
- Socioeconomic status (proxied by percent free/reduced lunch) is strongly associated with school-level ACT performance in this dataset. The estimated relationship is statistically significant and practically meaningful: modest improvements in SES correspond to measurable gains in average ACT scores.
- The model explains a large share of the variation in ACT (adj. R² ≈ 0.605), suggesting SES and its correlates are important predictors at the school level.

**Limitations**
- **Observational design:** This analysis is correlational; we cannot claim causal effects without stronger identification (panel data, IVs, or quasi-experiments).
- **Omitted variables:** School resources, teacher quality, student selection, and other community factors may confound the SES–ACT association.
- **Residual distribution:** Diagnostic tests (Omnibus, Jarque-Bera) indicate non-normal residuals and heavy tails. We used robust standard errors (HC1) to address heteroskedasticity, but non-normality suggests caution in inference about extreme predictions.
- **Measurement:** The SES proxy (`pct_free_lunch`) is an imperfect measure of socioeconomic context; measurement error would bias coefficient magnitudes.
